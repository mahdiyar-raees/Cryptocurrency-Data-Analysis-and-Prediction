```python
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
import time
import pandas as pd
from tabulate import tabulate




driver = webdriver.Chrome()
driver.get('https://coinmarketcap.com/historical/20230825/')

y = 500
for timer in range(0,20):
     driver.execute_script("window.scrollTo(0, "+str(y)+")")
     y += 1000  
     time.sleep(1)


rows = driver.find_elements(By.CSS_SELECTOR, ".cmc-table-row")
data = []
i = 0


for row in rows:
    driver.execute_script("arguments[0].scrollIntoView();",row)
    name = row.find_element(By.CSS_SELECTOR, ".cmc-table__column-name--name").text
    symbol = row.find_element(By.CSS_SELECTOR, ".cmc-table__cell--sort-by__symbol div").text
    rank = row.find_element(By.CSS_SELECTOR, ".cmc-table__cell--sort-by__rank").text
    main_link = row.find_element(By.CSS_SELECTOR, ".cmc-link").get_attribute("href")
    historical_link = main_link + "historical-data/"
    data.append([rank,name,symbol,main_link,historical_link])
data




df = pd.DataFrame(data, columns=['Rank', 'Name', 'Symbol', 'MainLink', 'HistoricalLink'])

styled_df = df.style.set_table_styles([
    {'selector': 'th', 'props': [('background-color', 'blue'), ('color', 'white')]},
    {'selector': 'td', 'props': [('background-color', 'white'), ('color', 'black')]},
    {'selector': 'table', 'props': [('border-collapse', 'collapse')]},
    {'selector': 'th, td', 'props': [('border', '1px solid black')]}
]).hide_index()
display(styled_df)
df.to_csv('D:/data_scienc_quera/firstTable.csv', index=False)

# for d in data:
#     driver.get(d[4])
#     button_locator = (By.XPATH, "//button[contains(@class, 'sc-16891c57-0 dalfmx')]")
#     button = WebDriverWait(driver, 10).until(EC.element_to_be_clickable(button_locator))

#     # Click the button
#     button.click()


#     # Wait for the div element to be visible
#     div_locator = (By.XPATH, "//div[@class='sc-5259c8b2-4 heoICr']")
#     div_element = WebDriverWait(driver, 10).until(EC.visibility_of_element_located(div_locator))

#     # Find the "Last 365 days" button within the div element
#     button_locator = (By.XPATH, ".//li[text()='Last 365 days']")
#     li_element = div_element.find_element(*button_locator)

#     # Click the "Last 365 days" button
#     driver.execute_script("arguments[0].click();", li_element)


#     buttons = driver.find_element_by_xpath("//button[@data-sensors-click='true' and text()='Continue']")
#     y = 100
#     for timer in range(0,2):
#          driver.execute_script("window.scrollTo(0, "+str(y)+")")
#          y += 150
#          time.sleep(3)
#     buttons.click()
#     for timer in range(0,2):
#          driver.execute_script("window.scrollTo(0, "+str(y)+")")
#          y -= 150
#          time.sleep(1)


#     buttons = driver.find_elements_by_xpath("//button[contains(@class, 'sc-16891c57-0 dalfmx')]")

#     buttons[1].click()
    




```
